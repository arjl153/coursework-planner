# Get courses list from html

```python
from bs4 import BeautifulSoup
```

```python
def get_courses_dict(file):
    with open(file, 'r') as file:
        html = file.read()

    soup = BeautifulSoup(html, 'html.parser')
    li_tags = soup.find_all('li')
    course_data = []
    
    for li in li_tags:
        url = li.a['href']
        dept, course_id, course_title = li.a.text.split(' ', 2)
        course = {
            'dept': dept,
            'url': url,
            'course_id': course_id,
            'course_title': course_title
        }
        course_data.append(course)
    return course_data

def print_courses(course_data):
    for course in course_data:
        print(f"dept ID: {course['dept']}")
        print(f"Course ID: {course['course_id']}")
        print(f"Course Title: {course['course_title']}")
        print(f"Url: {course['url']}")
        print()

spring_courses = get_courses_dict('spring.html')
print_courses(spring_courses)
```

```python
fall_courses = get_courses_dict('fall.html')
print_courses(fall_courses)
```

## Save html files of urls

```python
import os
import requests
```

```python
def save_html_files(courses, semester):
    directory = "html_files/" + semester
    os.makedirs(directory, exist_ok=True)
    n = len(courses)
    for i, course in enumerate(courses):
        link = course['url']
        filename = link.split("/")[-1]
        response = requests.get(link)
        
        if response.status_code == 200:
            filepath = os.path.join(directory, f"{filename}.html")
            with open(filepath, "w") as file:
                file.write(response.text)
            course['filename'] = f"{filename}.html"
            print(f"Saved {filename}.html")
            print(f"Completed {i+1}/{n}")
            print()
        else:
            print(f"Failed to fetch {filename}")

save_html_files(spring_courses, 'spring')
```

## Parsing the html files for the corses

```python
def parse_html_files(semester, course_list):  
    directory = 'html_files/' + semester + '/'
    updated_courses = []
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        if os.path.isfile(file_path):
            course_dict = next((course for course in course_list if course['filename'] == filename), None)
            with open(file_path, 'r') as file:
                html_data = file.read()

            soup = BeautifulSoup(html_data, 'html.parser')
            filter_results_div = soup.find('div', id='filter-results')
            meetings = []
            class_wrappers = filter_results_div.find_all('div', class_='class-wrapper')
            course_desc = soup.find('div', class_='full-width').find('p').text
            
            for class_wrapper in class_wrappers:
#                 class details
                component = class_wrapper.find('td', text=True).text
                credits = class_wrapper.find('td', text=True).find_next('td').text
                status = class_wrapper.find('td', text=True).find_next('td').find_next('td').find_next('td').text
                time = class_wrapper.find('td', text=True).find_next('td').find_next('td').find_next('td').find_next('td').text
                day = class_wrapper.find('td', text=True).find_next('td').find_next('td').find_next('td').find_next('td').find_next('td').text
                facility = class_wrapper.find('td', text=True).find_next('td').find_next('td').find_next('td').find_next('td').find_next('td').find_next('td').text
                instructor = class_wrapper.find('td', text=True).find_next('td').find_next('td').find_next('td').find_next('td').find_next('td').find_next('td').find_next('td').text
                
                class_detail = class_wrapper.find('div', class_='class-detail')
                class_header = class_detail.find('div', class_='class-header')
                p_tags = class_header.find_all('p')
                course_type = p_tags[0].find_all('strong')[0].text
                mode = p_tags[0].find_all('strong')[1].text

                seats_info = p_tags[1].find_all('strong')

                total_seats = seats_info[0].text
                available_seats = seats_info[1].text
                waitlisted_seats = seats_info[2].text

                class_content_ignore = class_detail.find('div', class_='class-content-ignore')
                li_elements = class_content_ignore.find_all('li')

                notes = [li.text for li in li_elements] if li_elements else []

                meetings.append({
                    'component': component,
                    'credits': credits,
                    'status': status,
                    'time': time,
                    'day': day,
                    'facility': facility,
                    'instructor': instructor,
                    'course_type': course_type,
                    'mode': mode,
                    'total_seats': total_seats,
                    'available_seats': available_seats,
                    'waitlisted_seats': waitlisted_seats,
                    'notes': notes
                })
            course_dict['meetings'] = meetings
            course_dict['description'] = course_desc
            updated_courses.append(course_dict)
        else:
            print("Failed to read the file:", filename)
    return updated_courses

complete_spring_courses = parse_html_files('spring', spring_courses)
```
